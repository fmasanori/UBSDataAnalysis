```python
#Quantas e quais UBS eram localizadas em bairros centrais?
base[base.dsc_bairro == "CENTRO"]
```

```python
sjc = folium.Map(
    location=[-23.220586, -45.901895],
    zoom_start=12
)
```

```python
ubs_sjc = base[base['dsc_cidade'] == "São José dos Campos"]
```

```python
#Mapa das UBS em São José dos Campos
for _, base in ubs_sjc.iterrows():
    folium.Marker(
        location=[base['vlr_latitude'], base['vlr_longitude']],
    ).add_to(sjc)

sjc
```
